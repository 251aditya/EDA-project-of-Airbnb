#### Chart - 4

````python
# Chart - 4 visualization code for visualize which type of rooms are more demanded..!
preferableRoomType = Airbnb.groupby(['room_type'])['minimum_nights'].count().reset_index()
preferableRoomType = preferableRoomType.sort_values(by='minimum_nights').head(10)
preferableRoomType
````

````python
room_types = preferableRoomType['room_type']
nights = preferableRoomType['minimum_nights']
# creating the pie plot
explode = (0.1,0.1, 0.1) # explode is use for gape between the pie chart
colors = ('red','green', 'purple')
plt.pie(nights, autopct='%.2f%%',colors = colors,labels = room_types ,explode = explode,shadow=True)
plt.title('Traffic area based on minimum nights booked',color = 'brown')
plt.show()
````

##### 1. Why did you pick the specific chart?

* I picked this chart because I want to visualize which type of rooms are more demanded?
